# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import pyperclip
import time

# pyautogui.click -> clicar
# pyautogui.write -> escrever
# pyautogui.press -> pressionar
# pyatugui.hotkey -> atalho

pyautogui.PAUSE = 1

# Passo 1: Entrar no sistema da empresa (no link da nuvem NAV Brasil: https://navbrasilgovbr-my.sharepoint.com/:f:/r/personal/thiago_luiz_navbrasil_gov_br/Documents/Aulas/Python)
# pyautogui.press("win")
# pyautogui.write("chrome")
# pyautogui.press("enter")
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://navbrasilgovbr-my.sharepoint.com/:f:/r/personal/thiago_luiz_navbrasil_gov_br/Documents/Aulas/Python")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(5)

# Passo 2: Navegar até o Local do relatório (entrar na pasta Exportar)
# usar pyautogui.position() para atualizar a posição do mouse
pyautogui.click(x=385, y=352, clicks=1, button="right")
time.sleep(2)

# Passo 3: Exportar o relatório (fazer o download)
pyautogui.click(x=281, y=353, button="right")
pyautogui.click(x=619, y=353, button="right")
pyautogui.click(x=665, y=409, button="right")
pyautogui.press("enter")
time.sleep(6)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4: Calcular os indicadores (faturamento e quantidade de produtos)
# pandas deve ser importado
# numpy e openpyxl já estão no pacote Jupyter
# tabula (converter pdf em pandas)
# possível abrir o Excel online, ou seja, diretamente da nuvem
import pandas as pd

tabela = pd.read_excel(r"C:\Users\basto\Desktop\Vendas - Dez.xlsx")
display(tabela)

faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
print(faturamento)
print(quantidade)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 5: Enviar um e-mail para a diretoria

# abrir aba e entrar no e-mail
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(5)

# clicar no botão escrever
pyautogui.click(x=100, y=206, button="right")

# preencher as informações do e-mail
# destinatário
pyautogui.write("thibastos@live.com")
pyautogui.press("tab") #seleciona o e-mail
pyautogui.press("tab") #pula para o campo de assunto

# assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") #pula para o campo corpo do e-mail

# corpo do e-mail
# o f antes do texto serve para inserir variáveis
# int - inteiro
# float - casa decimal - f = número de casas decimais
texto = f"""Prezados, bom dia!

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Atenciosamente,
Thiago Bastos."""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# enviar e-mail utilizando o atalho do Gmail
pyautogui.hotkey("ctrl", "enter")
```
