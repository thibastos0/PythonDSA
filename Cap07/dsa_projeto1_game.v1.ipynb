```python
# Projeto 1 - Desenvolvimento de Game em Linguagem Python - Versão 1

# Import
import random
from os import system, name

# Função para limpar a tela a cada execução
def limpa_tela():
	# windows
	if name == 'nt':
		_= system('cls') # retorno do comando descartado em '_'

	# linux ou mac
	else:
		_= system('clear')

# jogo
def game():

	limpa_tela()

	print("\nBem-vindo(a) ao jogo da forca!")
	print("Adivinhe a palavra abaixo:\n")

	#lista de palavras para o jogo
	palavras = ['BANANA', 'ABACATE', 'UVA', 'MORANGO', 'LARANJA']

	# escolha aleatória da palavra
	palavra = random.choice(palavras)

	# List comprehension
	letras_descobertas = ['_' for letra in palavra]

	# número de chances
	chances = 6

	# lista para as letras erradas
	letras_erradas = []

	# loop número de chances
	while chances > 0:

		print(" ".join(letras_descobertas)) # join: junção de palavras apenas
		print(f"\nChances restantes: {chances}")
		print(f"Letras erradas: {' '.join(letras_erradas)}")

		# Tentativa
		tentativa = input("\nDigite uma letra: ").upper()


		# Verifica letra
		if tentativa in palavra:
			for i, letra in enumerate(palavra):
				if tentativa == letra:
					letras_descobertas[i] = letra
		else:
			chances -= 1
			letras_erradas.append(tentativa)

		# fim de jogo com vitória

		if "_" not in letras_descobertas:
			print(f"\nVocê venceu, a palavra era: {palavra}")
			break

	if "_" in letras_descobertas:
		print(f"\nVocê perdeu, a palavra era: {palavra}")
```

```python
# Bloco main
if __name__ == "__main__":
  game()
  print("\n\033[mParabéns. Você está aprendendo programação em Python com a DSA. :)\n")
```
